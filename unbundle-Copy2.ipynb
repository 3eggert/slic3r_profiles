```python
from docopt import docopt
import os
import configparser
import hashlib
```

```python
doc = """Slic3r Config Unbundler

Usage:
  unbundle.py <config_bundle>

Options:
  -h --help     Show this screen.
"""

arguments = docopt(doc, version='Naval Fate 2.0')
print(arguments)
```

```python
bundle_config = "Slic3r_config_bundle.ini"
```

```python
def unbundle(bundle_config=None):
    if bundle_config is None:
        bundle_config = "Slic3r_config_bundle.ini"

    config = configparser.ConfigParser(strict=False, interpolation=None)
    config.read(bundle_config)
```

```python
def unbundle_section(config, cfg_section):
    setting_type, setting_name = cfg_section.split(":")
    if not os.path.exists(setting_type):
        os.mkdir(setting_type)
    setting_file = os.path.join(setting_type, setting_name)

    header_cfg = dict()

    filehash = hashlib.sha256()
    filehash.update(open(bundle_config, 'rb').read())
    header_cfg["sha256"] = filehash.hexdigest()


    header_cfg["bundle_config"]=bundle_config
    header_cfg["setting_type"]=setting_type.capitalize()
    header_cfg["setting_name"]=setting_name

    ini_header = """# {setting_type} Setting: {setting_name}
#
# This file is automatically generated.
# Permanent changes should be made in '{bundle_config}'
# SHA256 = {sha256}
""".format(**header_cfg)

    ini_tmp = list()
    for setting, value in config[cfg_section].items():
        ini_tmp.append("{} = {}".format(setting, value))

    ini_body = "\n".join(ini_tmp)
    
    
    with open(setting_file, 'w') as fid:
        print(ini_header, file = fid)
        print(ini_body, file=fid)

for cfg_section in config.sections():
    if ":" in cfg_section:
        unbundle_section(config=config, cfg_section=cfg_section)
```

```python
unbundle(bundle_config=bundle_config)
```
